```python
import pandas as pd
import numpy as np
```

##### 데이터프레임 불러오기

```python
# Data3.csv
df1=pd.read_csv('Data3.csv')
df1.head()
```

##### 필요없는 columns 정리하기

```python
df2=df1[
    ['DEMAND','DA_CC','DA_EC','DA_LMP','DA_MLC','DewPnt','DryBulb','RT_CC',
    'RT_EC','RT_LMP','RT_MLC']
]
df2.head()
```

##### dataset 분리하기

```python
from sklearn.model_selection import train_test_split
```

```python
x=df2[['DA_CC','DA_EC','DA_LMP','DA_MLC','DewPnt','DryBulb','RT_CC',
    'RT_EC','RT_LMP','RT_MLC']]
y=df2[['DEMAND']]
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

##### 모델 생성

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

##### 예측하기

```python
dfP=pd.read_csv('Predictor3.csv')
dfP.head()
```

```python
dfP2=dfP[
    ['DA_CC','DA_EC','DA_LMP','DA_MLC','DewPnt','DryBulb','RT_CC',
    'RT_EC','RT_LMP','RT_MLC']
]
dfP2.head()
```

```python
from tqdm.notebook import tqdm

predictList=[]
for i in tqdm(range(len(dfP2))):
    my_data=[[dfP2.iloc[i].DA_CC,
              dfP2.iloc[i].DA_EC,
              dfP2.iloc[i].DA_LMP,
              dfP2.iloc[i].DA_MLC,
              dfP2.iloc[i].DewPnt,
              dfP2.iloc[i].DryBulb,
              dfP2.iloc[i].RT_CC,
              dfP2.iloc[i].RT_EC,
              dfP2.iloc[i].RT_LMP,
              dfP2.iloc[i].RT_MLC,
             ]]
    my_predict = mlr.predict(my_data)
    predictList.append(my_predict.tolist()[0][0])
```

```python
predictList
```

```python
dfP.head()
```

```python
dfP['DEMAND']=predictList
```

```python
dfP.head()
```

```python
dfP.to_csv('Predict3.csv')
```
