```python
iris_txt_path=open('c://python//iris.txt','rt')
iris_txt=iris_txt_path.readlines()
for row in iris_txt:
    print(row)
```

```python
import csv
with open('c://python//iris.csv','w',newline='') as iris_csv:
    writer=csv.writer(iris_csv,delimiter=',')
    for row in iris_txt:
        writer.writerow(row)
```
