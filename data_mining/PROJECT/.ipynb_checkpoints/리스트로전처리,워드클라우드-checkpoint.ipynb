# 워드클라우드

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image
import numpy as np
```

```python
alice_mask = np.array(Image.open("melon2.png"))
```

```python
def makeWordCloud(List):
    cnt= Counter(List)
    Dict = dict(cnt.most_common())
    wordcloud = WordCloud(font_path='font/NanumGothic.ttf', background_color='white').generate_from_frequencies(Dict)    
    plt.figure(figsize=(15,15)) # 이미지 사이즈 지정
    plt.imshow(wordcloud, interpolation='bilinear') # 이미지의 부드럽기 정도
    plt.axis('off') # x y 축 숫자 제거
    plt.show()
```

```python
def makeWordCloud_mask(List):
    cnt= Counter(List)
    Dict = dict(cnt.most_common())
    wordcloud = WordCloud(font_path='font/NanumGothic.ttf', background_color='white',mask = alice_mask).generate_from_frequencies(Dict)    
    plt.figure(figsize=(15,15)) # 이미지 사이즈 지정
    plt.imshow(wordcloud, interpolation='bilinear') # 이미지의 부드럽기 정도
    plt.axis('off') # x y 축 숫자 제거
    plt.show()
```

### 장르 WordCloud

```python
makeWordCloud(genreList_split)
```

### 곡분류 WordCloud

```python
makeWordCloud(labelList_split)
```

### 가수 분류 WordCloud

```python
makeWordCloud(singertypeList_full)
```
