```python
import pandas as pd
df = pd.read_excel('melon100_labeling_excel.xlsx',index_col='순위')

# 곡분류 (2020년일때 주석해제)
labelList_full=list(df['곡분류'])
labelList_split=[]
for i in range(100):
    labelList_split.extend((labelList_full[i].split(',')))

# 가수분류
singertypeList_full=list(df['가수분류'])
singertype_unique=list(set(singertypeList_full))

# 장르
genreList_full=list(df['장르'])
genreList_split=[]
for i in range(100):
    genreList_split.extend((genreList_full[i].split(',')))
genreList_unique=list(set(genreList_split))

from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter


def makeWordCloud(List):
    cnt= Counter(List)
    Dict = dict(cnt.most_common())
    wordcloud = WordCloud(font_path='c:/python/NanumGothic.ttf', background_color='white').generate_from_frequencies(Dict)    
    plt.figure(figsize=(15,15)) # 이미지 사이즈 지정
    plt.imshow(wordcloud, interpolation='bilinear') # 이미지의 부드럽기 정도
    plt.axis('off') # x y 축 숫자 제거
    plt.show()

makeWordCloud(genreList_split)
# 2020년일때 주석해제
makeWordCloud(labelList_split)
makeWordCloud(singertypeList_full)
```
